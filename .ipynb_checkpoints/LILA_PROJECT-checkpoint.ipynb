```python
import statsmodels.api as sm
```

```python
SURABAYA = data[data['Branch'] == 'SURABAYA']
JAKARTA = data[data['Branch'] == 'JAKARTA']

n_SBY = len(SURABAYA)
mean_SBY = SURABAYA['Booking NTF Amount'].mean()
std_SBY = SURABAYA['Booking NTF Amount'].std()

n_JKT = len(JAKARTA)
mean_JKT = JAKARTA['Booking NTF Amount'].mean()
std_JKT = JAKARTA['Booking NTF Amount'].std()

print(n_SBY, mean_SBY, std_SBY)
print(n_JKT, mean_JKT, std_JKT)
```

```python
print (('zstat', 'p-value'))
sm.stats.ztest(SURABAYA['Booking NTF Amount'], JAKARTA['Booking NTF Amount'], alternative='two-sided')
```
