```python
import numpy as np
import pandas as pd
import math
import statistics
import seaborn as sns
import statistics as stats
import scipy.stats
from scipy.stats import uniform
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot'])
%matplotlib inline
```

```python
#Menghilangkan Kolom Yang Tidak Perlu#
data =df.drop(['Semester', 'Quarter', 'Customer Industry Type'], axis=1)
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['Booking NTF Amount'] = pd.to_numeric(data['Booking NTF Amount'], errors='coerce').convert_dtypes()
```

```python
data.head(5)
```

```python
#Mengganti Nama Kolom#

data.rename(columns={'Booking NTF Amount': 'Booking NTF', 'Industry Type Group': 'Industry Type'}, inplace=True)
data
```

```python
data['Booking NTF'] = data['Booking NTF'].astype(np.int64)
```

```python
data.info()
```

```python
#Menghapus Missing Data#
data = data.dropna()
data.head(5)
```

```python
#Sortir Booking NTF < Rp.500 Juta##

data1 = data[data['Booking NTF'] > 500000000]
```

```python
data1.info()
```

```python
#Pengelompokan Booking NTF Per Cabang#
BokCbg = data.groupby('Branch')
```

```python
BokNTF = data1.groupby(['Branch', 'Year'])['Booking NTF'].sum().reset_index()
BokNTF
```

```python
#Mengubah Baris Tahun menjadi Kolom#
BokNTF1 = BokNTF.reset_index().pivot(columns= 'Year', index= 'Branch', values='Booking NTF')
BokNTF1
```

```python
BokNTF1=BokNTF1.fillna(0)
```

```python
# menjumlahkan Booking ntf setiap tahun #
BokNTF1['Total'] = BokNTF1.sum(axis=1)
```

```python
# sorting total booking  secara descending(besar ke kecil) #
BokNTF1.sort_values(by='Total', ascending=False, axis=0, inplace=True)
```

```python
BokNTF1top5 = BokNTF1.head(5)
```

```python
BokNTF1top5
```

```python
# transpose data frame
BokNTF1top5t = BokNTF1top5.T
BokNTF1top5t
```

```python
BokNTF1top = BokNTF1top5t.drop('Total')
BokNTF1top
```

```python
BokNTF1top.plot(kind='area',
stacked=False,
figsize=(20,10),
alpha=0.25,
   )
plt.title('Top 5 Branch Contributed 2017-2021')
plt.ylabel('Booking NTF')
plt.xlabel('Years')

plt.show()
```
